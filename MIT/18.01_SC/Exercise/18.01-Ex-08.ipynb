### 问题
- 1、$\sin{x}$ 是一个周期性函数，而 $\frac{1}{x}$ 是双曲线函数且在第一、三象限内分别单调递减  
2、结合来看 $\frac{\sin{x}}{x}$ 是一个偶函数，在 $x=0$ 点无定义且 $\lim_{x \to 0}\frac{\sin{x}}{x}=1$  
3、函数 $\frac{\sin{x}}{x}$ 除在 $x=0$ 外，与 $x$ 轴的交点与 $\sin{x}$ 相同  
4、以 $x$轴正方向为例，可以将 $\frac{\sin{x}}{x}$ 看作一个振幅为 $\frac{1}{x}$ 的震荡函数，且这个振幅沿 $x$ 轴方向越来越小
