### 问题1
- 函数作用域为 $[0, \infty)$，只能使用右极限  
$\lim_{x \to 0^+}\sqrt{x}=0$

### 问题2
- 这是一个双曲线，在 $x=-1$ 处曲线不连续且伸展方向相反，需要使用单侧极限  
  
$\lim_{x \to -1^-}\frac{1}{x+1}=-\infty$  
$\lim_{x \to -1^+}\frac{1}{x+1}=\infty$

### 问题3
- 曲线在 $x=1$ 处虽不连续但伸展方向相同，可以使用双侧极限  
  
$\lim_{x \to 1}\frac{1}{(x-1)^4}=\infty$

### 问题4
- 函数在 $x=0$ 处连续，可以使用双侧极限  
  
$\lim_{x \to 0}|sinx|=0$

### 问题5
- 函数在 $x=0$ 处为 JUMP Discontinuty，需要使用单侧极限  
  
$\lim_{x \to 0^-}\frac{|x|}{x}=-1$  
$\lim_{x \to 0^+}\frac{|x|}{x}=1$
